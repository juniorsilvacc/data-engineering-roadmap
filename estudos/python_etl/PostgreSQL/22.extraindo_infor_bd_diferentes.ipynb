```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
server = 'localhost'
database = 'python_etl_varias_bases'
user = 'postgres'
password = 'root'

try:
    engine = create_engine(f'postgresql://{user}:{password}@{server}/{database}')
    print("Conexão bem-sucedida ao PostgreSQL!")
except Exception as e:
    print("Erro ao conectar ao PostgreSQL:", e)
```

```python
query = """
    SELECT
        *
    FROM items
"""

items = pd.read_sql(query, engine)
items.head(5)
```

Criar conexão com o SEGUNDO banco de dados

```python
server2 = 'localhost'
database2 = 'python_etl'
user = 'postgres'
password = 'root'

try:
    engine2 = create_engine(f'postgresql://{user}:{password}@{server2}/{database2}')
    print("Conexão bem-sucedida ao PostgreSQL!")
except Exception as e:
    print("Erro ao conectar ao PostgreSQL:", e)
```

```python
query = """
    SELECT
        *
    FROM produtos
"""

produtos = pd.read_sql(query, engine2)
produtos.head(5)
```

```python
join = pd.merge(items, produtos, left_on="product_id", right_on="id", how="left")
```
