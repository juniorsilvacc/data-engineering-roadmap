```python
import tkinter as tk
import sqlite3
import tkinter.messagebox as messagebox
import pandas as pd

# Criar banco
def criar_banco():
    conexao = sqlite3.connect("cadastro_produtos.db")
    cursor = conexao.cursor()
    cursor.execute("""
            CREATE TABLE IF NOT EXISTS produtos (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                valor REAL,
                fornecedor TEXT,
                categoria TEXT,
                responsavel_cadastro TEXT,
                data_criacao DATETIME DEFAULT CURRENT_TIMESTAMP
            )
        """)

# Função para salvar os dados no banco
def salvar_dados():
    try:
        nome = entry_nome.get()
        valor = entry_valor.get()
        fornecedor = entry_fornecedor.get()
        categoria = entry_categoria.get()
        responsavel_cadastro = entry_responsavel.get()

        # Validação básica
        if not (nome and valor):
            messagebox.showwarning("Campos obrigatórios", "Preencha o nome e valor do produto.")
        
        try:
            valor_float = float(valor)
        except Exception as e:
            messagebox.showerror("Erro de Valor", "Digite o número válido para o valor.")
            return

        # Salvar
        conexao = sqlite3.connect("cadastro_produtos.db")
        cursor = conexao.cursor()
        cursor.execute("""
                INSERT INTO produtos (nome, valor, fornecedor, categoria, responsavel_cadastro)
                VALUES (?, ?, ?, ?, ?)
            """, (nome, valor_float, fornecedor, categoria, responsavel_cadastro))
        
        conexao.commit()
        conexao.close()

        messagebox.showinfo("Sucesso", "Produto salvo com sucesso")

        # Limpar campos após salvar
        entry_nome.delete(0, tk.END)
        entry_valor.delete(0, tk.END)
        entry_fornecedor.delete(0, tk.END)
        entry_categoria.delete(0, tk.END)
        entry_responsavel.delete(0, tk.END)
    except Exception as e:
        messagebox.showerror("Erro", f"Ocorreu um erro ao salvar: {e}")

# Criar janela
janela = tk.Tk()
janela.title('Cadastro de Produto')
janela.geometry('450x350') # Tamanho fixo da janela

# Frame central para organizar melhor
frame = tk.Frame(janela, padx=20, pady=20)
frame.pack(padx=10, pady=10)

# Gerar excel
def exportar_excel():
    conexao = sqlite3.connect("cadastro_produtos.db")
    df = pd.read_sql_query("SELECT * FROM produtos", conexao)
    conexao.close()

    df.to_excel("produtos.xlsx", index=False)
    messagebox.showinfo("Exportado", "Arquivo Excel criado com sucesso!")

# Função auxiliar pra criar campos
def criar_campo(label_text, row):
    label = tk.Label(frame, text=label_text, font=("Arial", 12, "bold"))
    label.grid(row=row, column=0, sticky="e", padx=10, pady=5)
    
    entry = tk.Entry(frame, width=35, font=("Arial", 10))
    entry.grid(row=row, column=1, padx=10, pady=5)
    
    return entry

# Criar os campos
entry_nome = criar_campo("Nome do Produto:", 1)
entry_valor = criar_campo("Valor do Produto:", 2)
entry_fornecedor = criar_campo("Fornecedor:", 3)
entry_categoria = criar_campo("Categoria:", 4)
entry_responsavel = criar_campo("Responsável:", 5)

# Botão de salvar
btn_salvar = tk.Button(frame, text="Salvar", bg="#4CAF50", fg="#000", width=20, font=("Arial", 12, "bold"), command=salvar_dados)
btn_salvar.grid(row=6, column=0, columnspan=2, pady=20)

# Botão de exportar
btn_salvar = tk.Button(frame, text="Exportar", bg="#4CAF50", fg="#000", width=20, font=("Arial", 12, "bold"), command=exportar_excel)
btn_salvar.grid(row=7, column=1, columnspan=2, pady=20)

# Iniciar janela e criar banco
criar_banco()
janela.mainloop()
```
