```python
import pandas as pd
import os
```

```python
v_2021 = "/home/juniorsilvadev/engenharia-dados/python-etl/Origem/Origem/arquivos_excel/Consolidar/Vendas2021.xlsx"
v_2022 = "/home/juniorsilvadev/engenharia-dados/python-etl/Origem/Origem/arquivos_excel/Consolidar/Vendas2022.xlsx"
v_2023 = "/home/juniorsilvadev/engenharia-dados/python-etl/Origem/Origem/arquivos_excel/Consolidar/Vendas2023.xlsx"
```

```python
df2021 = pd.read_excel(v_2021)
df2022 = pd.read_excel(v_2022)
df2023 = pd.read_excel(v_2023)
```

```python
df2021.head(5)
```

```python
# Extrai o nome do arquivo sem o caminho e a extensão usando o módulo os
nome_arquivo01 = os.path.splitext(os.path.basename(v_2021))[0]
nome_arquivo02 = os.path.splitext(os.path.basename(v_2022))[0]
nome_arquivo03 = os.path.splitext(os.path.basename(v_2023))[0]

# Adiciona o nome do arquivo
df2021['Nome do Arquivo'] = nome_arquivo01
df2022['Nome do Arquivo'] = nome_arquivo02
df2023['Nome do Arquivo'] = nome_arquivo03
```

```python
df2023.head(5)
```

```python
consolidado = pd.concat([df2021, df2022, df2023], ignore_index=True)
```

```python
pasta_destino = "/home/juniorsilvadev/engenharia-dados/python-etl/Origem/Origem/arquivos_excel/Consolidar/Concat/"

nome_arquivo = "ConsolidadoArquivosComNome.xlsx"

dados = []

caminho_completo = os.path.join(pasta_destino, nome_arquivo)

if os.path.exists(caminho_completo):
    df_existente = pd.read_excel(caminho_completo)
    dados.append(df_existente)

consolidado.to_excel(caminho_completo, index=False) 
```
