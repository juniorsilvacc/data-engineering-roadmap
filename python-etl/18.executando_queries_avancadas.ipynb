```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# O erro ocorre porque o pd.read_sql do Pandas não suporta diretamente objetos de conexão do psycopg2
```

```python
server = 'localhost'
database = 'python_etl_varias_bases'
user = 'postgres'
password = 'root'

try:
    engine = create_engine(f'postgresql://{user}:{password}@{server}/{database}')
    print("Conexão bem-sucedida ao PostgreSQL!")
except Exception as e:
    print("Erro ao conectar ao PostgreSQL:", e)
```

```python
query = """
    SELECT
        t1.id AS id_venda,
        t1.total_price AS total_preco,
        t3.name AS nome_produto,
        t4.name AS categoria,
        t5.first_name AS cliente,
        t5.email,
        t5.cell_phone AS telefone,
        t2.status,
        t2.created_at
    FROM items AS t1
        LEFT JOIN ordens AS t2 ON t1.order_id = t2.id
        LEFT JOIN produtos AS t3 ON t1.product_id = t3.id
        LEFT JOIN categorias AS t4 ON t3.id_category = t4.id
        LEFT JOIN clientes as t5 ON t2.customer_id = t5.id
"""

items = pd.read_sql(query, engine)
items.head(10)
```
