```python
diretorio = ("/home/juniorsilvadev/engenharia-dados/03-python-etl-do-basico-ao-avancado/Origem/Origem/arquivos_excel/Consolidar")

# Lista de dados vazia para armazenar os dados
dados = []

# Iterar sobre os arquivos do excel
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_excel(caminho_arquivo)
        dados.append(df)

# Consolidar em um único Dataframe
consolidado = pd.concat(dados)
```

Salvando consolidado em uma pasta em formato Excel

```python
pasta_destino = ("/home/juniorsilvadev/engenharia-dados/03-python-etl-do-basico-ao-avancado/arquivos-consolidados/")

nome_arquivo = 'VendasGeral.xlsx'

caminho_arquivo = pasta_destino + nome_arquivo

consolidado.to_excel(caminho_arquivo, index=False)
```

Salvando consolidado em uma pasta em formato CSV

```python
pasta_destino = ("/home/juniorsilvadev/engenharia-dados/03-python-etl-do-basico-ao-avancado/arquivos-consolidados/")

nome_arquivo = 'VendasGeral.csv'

caminho_arquivo = pasta_destino + nome_arquivo

consolidado.to_csv(caminho_arquivo, index=False, sep=',')
```

Verificar se arquivo já existe e excluir

```python
pasta_destino = ("/home/juniorsilvadev/engenharia-dados/03-python-etl-do-basico-ao-avancado/arquivos-consolidados/")

nome_arquivo = 'VendasGeral.xlsx'

caminho_completo = pasta_destino + nome_arquivo

# Verificar se o arquivo Excel já existe
if os.path.exists(caminho_completo):
    print(f"Arquivo existente encontrado: {caminho_completo}, será removido.")
    os.remove(caminho_completo)

# Garantir que o DataFrame consolidado existe e não está vazio
if consolidado.empty:
    print("O DataFrame 'consolidado' está vazio. Nenhum arquivo será criado.")
    exit()

# Criar o arquivo Excel
try:
    consolidado.to_excel(caminho_completo, index=False)
    print(f"Novo arquivo criado em: {caminho_completo}")
except Exception as e:
    print(f"Erro ao criar o arquivo: {e}")
```
